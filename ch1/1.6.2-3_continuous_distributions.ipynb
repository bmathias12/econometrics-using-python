## PDF - Normal

```python
# support of normal density
x_range = np.linspace(-4,4, num=100)

# PDF for all these values
pdf = stats.norm.pdf(x_range)
```

```python
plt.plot(x_range, pdf, color='k')
plt.xlabel('x')
plt.ylabel('dx')
```

## CDF

For all distributions the CDF $F(x) = P(X <= x)$ represents the probability that the random variable X takes a value of *at most* x

```python
stats.norm.cdf(1.96) - stats.norm.cdf(-1.96)
```

```python
stats.norm.cdf(1.96)
```

## Plots of the CDF of Discrete and Continuous RV

```python
# binomial
x_binom = np.linspace(-1, 10, num=1000)
cdf_binom = stats.binom.cdf(x_binom, 10, 0.200)

plt.step(x_binom, cdf_binom, color='k')
plt.xlabel('x')
plt.ylabel('Fx')
```

```python
# normal
x_norm = np.linspace(-4, 4, num=1000)
cdf_norm = stats.norm.cdf(x_norm)

plt.plot(x_norm, cdf_norm, color='k')
```

## Quantile Function
The q-quantile $x[q]$ of a random variable is the value for which the probability to sample a value $x <= x[q]$ is just q. E.g., Given X is standard normal, the 0.975-quantile is $x[0.975] \approx 1.96$. So the probability to sample a value less or equal to 1.96 is 97.5%.

```python
stats.norm.ppf(0.975)
```
