```python
import wooldridge as woo
import statsmodels.formula.api as smf
import pandas as pd
import matplotlib.pyplot as plt

ceosal1 = woo.dataWoo('ceosal1')
```

```python
reg = smf.ols(formula='salary ~ roe', data=ceosal1)
results = reg.fit()
```

```python
print(results.summary())
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.scatter('roe', 'salary', data=ceosal1, marker='o', facecolors='none', edgecolors='red', s=20)
plt.plot(ceosal1['roe'], results.fittedvalues, color='black')
plt.ylabel('salary')
plt.xlabel('roe')
```

```python
vote1 = woo.dataWoo('vote1')
```

```python
reg = smf.ols(formula='voteA ~ shareA', data=vote1)
results = reg.fit()
```

```python
print(results.summary())
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.scatter('shareA', 'voteA', data=vote1, marker='o', facecolors='none', edgecolors='red', s=20)
plt.plot(vote1['shareA'], results.fittedvalues, color='black')
plt.ylabel('voteA')
plt.xlabel('shareA')
```
