```python
import wooldridge as woo
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Semi-Log Model

```python
wage1 = woo.dataWoo('wage1')
```

```python
# Estimate log-level model
reg = smf.ols(formula='np.log(wage) ~ educ', data=wage1)
results = reg.fit()
```

```python
print(results.summary())
```

```python
x = wage1['educ']
y = np.log(wage1['wage'])
data = wage1

def plot_ols(x, y, data, results):
    fig, ax = plt.subplots(figsize=(10,6))
    plt.scatter(x, y, marker='o', facecolors='none', edgecolors='red', s=20)
    plt.plot(x, results.fittedvalues, color='black')
    plt.ylabel(y.name)
    plt.xlabel(x.name)
    plt.show()

plot_ols(x, y, data, results)
```

## Log-Log Model

```python
ceosal1 = woo.dataWoo('ceosal1')

reg = smf.ols(formula='np.log(salary) ~ np.log(sales)', data=ceosal1)
results = reg.fit()
```

```python
print(results.summary())
```

```python
x = np.log(ceosal1['sales'])
y = np.log(ceosal1['salary'])
data = ceosal1


plot_ols(x, y, data, results)
```
